### Drop other languages

```python
from langdetect import detect
```

```python
#from deep_translator import GoogleTranslator
#from deep_translator import MyMemoryTranslator
```

```python
indexes = []
for i in range(0, len(df["book_desc"])+1):
    try:
        if detect(str(df["book_desc"][i]))!= 'en':
            indexes.append(df.index[i])
    except KeyError:
        pass
    except Exception as e:
        print(e)
        indexes.append(df.index[i])
```

```python
len(indexes)
```

```python
df = df.drop(indexes, axis=0)
```

```python
df.shape
```

```python
df.duplicated().value_counts()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['book_desc'])
```

```python
df.describe()
```

```python
df.image_url[0]
```

```python
df = df.drop(columns=["book_edition","book_format","book_isbn","book_review_count"])
```

```python
df.head()
```

```python
def listed(x):
    if type(x) != str:
        return x
    return x.split("|")
```

```python
df["book_authors"] = df["book_authors"].apply(listed)
```

```python
df["genres"] = df["genres"].apply(listed)
```

```python
df["genres"]
```

```python
def pages(x):
    if type(x)==str:
        if "pages" in x:
            return float(x[:-6])
        return float(x[:-5])
    return x
```

```python
df["book_pages"] = df["book_pages"].apply(pages)
```
