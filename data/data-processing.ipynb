```python
df = df.drop(columns=["book_edition","book_format","book_isbn","book_review_count"])
```

```python
df.head()
```

```python
def listed(x):
    if type(x) != str:
        return x
    return x.split("|")
```

```python
df["book_authors"] = df["book_authors"].apply(listed)
```

```python
df["genres"] = df["genres"].apply(listed)
```

```python
df["genres"]
```

```python
def pages(x):
    if type(x)==str:
        if "pages" in x:
            return float(x[:-6])
        return float(x[:-5])
    return x
```

```python
df["book_pages"] = df["book_pages"].apply(pages)
```
